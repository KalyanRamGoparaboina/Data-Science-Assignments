```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
basket = df.astype(bool)
basket.head()
```

```python
#Association rule mining
```

```python
frequent_itemsets = apriori(basket, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
#1.Lift is a measure that shows how much more likely the consequent of a rule is purchased when the antecedent is purchased, compared to its baseline probability.
#Confidence = Probability that consequent occurs given antecedent.
#Support of the consequent = Probability that consequent occurs at all in the dataset.

#why is it important in association rule
#Lift tells you the strength of association between products, beyond what would happen by chance.
#Lift = 1 â†’ Antecedent and consequent are independent (no association).
#Lift > 1 â†’ Positive association; buying the antecedent increases likelihood of buying the consequent.
#Lift < 1 â†’ Negative association; buying the antecedent reduces likelihood of buying the consequent.
```

```python
#2.
#Support
#It tells you the popularity or frequency of an item or combination of items
#Formula:
#Support(A)=NumberÂ ofÂ transactionsÂ containingÂ A/TotalÂ numberÂ ofÂ transactions

#Confidence
#Confidence measures how often the consequent occurs when the antecedent occurs.
#It tells you the reliability of the rule A â†’ B.
#Formula:
#Confidence(AÂ â†’Â B)=Support(AÂ âˆ©Â B)/Support(A)
```
