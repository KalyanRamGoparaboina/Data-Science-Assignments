```python
#A total of 15 print-heads were randomly selected and tested until failure. The durability of each print-head (in millions of characters) was recorded as follows:1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29

import numpy as np
import scipy.stats as stats
# Data: durability in millions of characters
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 
        1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
# Convert to numpy array
data = np.array(data)

# Sample size, mean, std
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)   # sample standard deviation
se = std / np.sqrt(n)        # standard error

# Degrees of freedom
df = n - 1

# t critical value for 99% CI
t_crit = stats.t.ppf(0.995, df)

# Margin of error
margin = t_crit * se

# Confidence interval
ci_lower = mean - margin
ci_upper = mean + margin

print("Sample size (n):", n)
print("Sample mean:", round(mean, 4))
print("Sample std (s):", round(std, 4))
print("Standard error (SE):", round(se, 4))
print("t critical value:", round(t_crit, 4))
print("Margin of error:", round(margin, 4))
print("99% Confidence Interval: (", round(ci_lower, 4), ",", round(ci_upper, 4), ")")
```

```python
#b. Build 99% Confidence Interval Using Known Population Standard DeviationIf it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for the mean number of characters printed before failure.
import numpy as np
import scipy.stats as stats

# Given data: Durability of 15 print-heads (in millions of characters)
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07,
        1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

# Convert to numpy array
data = np.array(data)

# Known population standard deviation
sigma = 0.2

# Sample size and mean
n = len(data)
mean = np.mean(data)

# Standard error using population std
se = sigma / np.sqrt(n)

# z critical value for 99% confidence interval (two-tailed)
z_crit = stats.norm.ppf(0.995)  # 99% → 0.5% in each tail

# Margin of error
margin = z_crit * se

# Confidence interval
ci_lower = mean - margin
ci_upper = mean + margin

# Print results
print("Sample size (n):", n)
print("Sample mean:", round(mean, 4))
print("Population standard deviation (σ):", sigma)
print("Standard error (SE):", round(se, 4))
print("z critical value:", round(z_crit, 4))
print("Margin of error:", round(margin, 4))
print("99% Confidence Interval: (", round(ci_lower, 4), ",", round(ci_upper, 4), ")")
```
