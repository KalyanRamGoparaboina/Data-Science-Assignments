```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
for col in num_cols:
    if df[col].isnull().sum() > 0:
        median_value = df[col].median()
        df[col].fillna(median_value, inplace=True)
        print(f"✅ Filled missing values in '{col}' with median: {median_value}")
```

```python
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)
        print(f"✅ Filled missing values in '{col}' with mode: {mode_value}")
```

```python
#One-Hot Encoding:
#Creates new binary columns (0 or 1) for each category.
#pros:
#No ordinal relationship — treats all categories equally
#Works well with linear models
#Maintains interpretability — each column represents one category
#Helps avoid misleading order assumptions

#Cons:
#Increases dimensionality (too many columns if categories are many)
#Can lead to sparse data (many zeros)
#Slower training for large datasets
#Not suitable for high-cardinality columns (hundreds of unique values)
```

```python
#Label Encoder
#Assigns a numeric label to each category.
#Pros:
#Simple and compact representation
#No increase in the number of columns
#Useful for tree-based algorithms (Decision Trees, Random Forests) that can handle numeric labels
#Efficient for high-cardinality features

#Cons:
#Implies a false ordinal relationship between categories
#Can mislead linear models
#May affect model performance when categories are nominal 
```

```python
def work_hours_category(hours):
    if hours < 35:
        return 'Part-time'
    elif 35 <= hours <= 45:
        return 'Full-time'
    else:
        return 'Overtime'
        df['work_hours_category'] = df['hours-per-week'].apply(work_hours_category)
```

```python
def age_group(age):
    if age < 25:
        return 'Young'
    elif 25 <= age < 45:
        return 'Adult'
    elif 45 <= age < 65:
        return 'Senior'
    else:
        return 'Elderly'

df['age_group'] = df['age'].apply(age_group)
```

```python
df.head()
```

```python
#Explanation:
#work_hours_category
#What it does: Converts hours-per-week into categories:
#< 35 → "Part-time"
#35–45 → "Full-time"
#> 45 → "Overtime"

#age_group
#What it does: Groups age into levels:
#Young (<25)
#Adult (25–44)
#Senior (45–64)
#Elderly (65+)
```

```python
#Feature Extraction
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(contamination=0.01, random_state=42)  # 1% expected outliers
outliers = iso.fit_predict(df[num_cols])
df['outlier'] = outliers
df_clean = df[df['outlier'] == 1].drop('outlier', axis=1)
df.shape
```

```python
df_clean.shape
```

```python
from sklearn.feature_selection import mutual_info_classif 
target_col = 'income'  
X = df_clean.drop(columns=[target_col], errors='ignore')
y = df_clean[target_col]
```

```python
from sklearn.preprocessing import LabelEncoder
X_encoded = X.copy()
for col in X_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))
```

```python
mi_scores = mutual_info_classif(X_encoded, y, discrete_features='auto', random_state=42)
mi_df = pd.DataFrame({'feature': X_encoded.columns, 'score': mi_scores})
mi_df = mi_df.sort_values(by='score', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='score', y='feature', data=mi_df, palette='Blues_r')
plt.title("Feature Importance using Mutual Information")
plt.xlabel("Mutual Information Score")
plt.ylabel("Feature")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import ppscore

pps_matrix = ppscore.matrix(df)
pps_matrix.head()
```

```python
pps_pivot = pps_matrix.pivot(index='x', columns='y', values='ppscore')

plt.figure(figsize=(12,8))
sns.heatmap(pps_pivot, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Predictive Power Score (PPS) Matrix")
plt.show()
```

```python
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix (Numerical Features)")
plt.show()
```
