```python
#Dimensionality Reduction with PCA
```

```python
print(df.columns.tolist())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
y = df.iloc[:, 0]
X = df.iloc[:, 1:]     
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)
print("Cumulative variance:", explained_variance.cumsum())
```

```python
plt.figure(figsize=(5,3))
plt.plot(range(1, len(explained_variance)+1), explained_variance.cumsum(), marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Explained Variance")
plt.grid()
plt.show()
```

```python
#Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance
```

```python
pca = PCA()
pca.fit(X_scaled)

explained_var = pca.explained_variance_ratio_
cum_var = explained_var.cumsum()

plt.figure(figsize=(8,5))
plt.bar(range(1, len(explained_var)+1), explained_var, alpha=0.7, label='Individual explained variance')
plt.plot(range(1, len(cum_var)+1), cum_var, marker='o', color='red', label='Cumulative explained variance')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot & Cumulative Explained Variance')
plt.grid(True)
plt.legend()
plt.show()
```

```python
for i, (ev, cv) in enumerate(zip(explained_var, cum_var), 1):
    print(f"PC{i}: Individual={ev:.3f}, Cumulative={cv:.3f}")
```

```python
pca_columns = [f'PC{i+1}' for i in range(pca.n_components_)]
df_pca = pd.DataFrame(X_pca, columns=pca_columns)
df_final = pd.concat([y.reset_index(drop=True), df_pca], axis=1)
print(df_final.head())
```

```python
pca = PCA(n_components=7)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Get feature names
feature_names = X.columns
loadings = pd.DataFrame(pca.components_,
                        columns=feature_names,
                        index=[f'PC{i+1}' for i in range(pca.n_components_)])

print("PCA Component Loadings:")
print(loadings)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(loadings, annot=True, cmap='coolwarm', center=0)
plt.title("PCA Component Loadings (Feature Contributions)")
plt.show()
```

```python
import os
import warnings
from sklearn.cluster import KMeans
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["LOKY_MAX_CPU_COUNT"] = "4" 

warnings.filterwarnings("ignore")
wcss = []
for k in range(1, 11):
    model = KMeans(
        n_clusters=k,
        init='k-means++',
        random_state=42,
        n_init='auto'
    )
    model.fit(X_scaled)
    wcss.append(model.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o', color='blue')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)
plt.show
```

```python
kmeans = KMeans(
    n_clusters=3,
    init='k-means++',
    random_state=42,
    n_init='auto'
)
clusters = kmeans.fit_predict(X_scaled)
df_clusters = df.copy()
df_clusters['Cluster'] = clusters

print("\nâœ… Clustering completed! Sample data:")
print(df_clusters.head())
```

```python
df_plot = pd.DataFrame({
    'PCA1': X_pca[:, 0],
    'PCA2': X_pca[:, 1],
    'Cluster': clusters
})

plt.figure(figsize=(8,6))
for cluster in df_plot['Cluster'].unique():
    plt.scatter(
        df_plot[df_plot['Cluster'] == cluster]['PCA1'],
        df_plot[df_plot['Cluster'] == cluster]['PCA2'],
        label=f'Cluster {cluster}'
    )

plt.title('K-Means Clustering Visualization (PCA 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init='auto')
clusters = kmeans.fit_predict(X_scaled)

sil_score = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(X_scaled, clusters)
print(f"Daviesâ€“Bouldin Index: {db_index:.3f}")
```

```python
#Similarities:
#Number of clusters: Both the original and PCA-transformed datasets produced the same number of clusters (k=3), as expected since we chose k based on the Elbow Method.
#Overall cluster structure: The clusters in PCA-transformed data roughly correspond to the same groups seen in the original data. Points that were together in the original dataset generally remain together after PCA.
#Cluster separation trend: Both datasets show reasonably well-separated clusters, indicating K-Means captures the underlying structure of the wine dataset.
```

```python
#Differences
#Cluster compactness: Clusters in the PCA-transformed data often appear more compact and slightly better separated. This is because PCA reduces noise and focuses on the directions of maximum variance.
#Silhouette Score & Davies-Bouldin Index:
#Typically, silhouette score increases slightly after PCA, suggesting points are more tightly grouped within clusters.
#Davies-Bouldin index decreases slightly, indicating clusters are more distinct after dimensionality reduction.
#Visualization clarity: In 2D PCA scatter plots, the cluster boundaries are easier to visualize compared to projecting the original high-dimensional data.
```

```python
#Impact of Dimensionality Reduction on Clustering
#Improved cluster separation
#PCA removes less informative dimensions and noise, leaving the most important variance directions.
#As a result, clusters in PCA-transformed space are often more compact and better separated than in the original high-dimensional data.

#Enhanced interpretability and visualization
#Reducing the data to 2â€“3 principal components allows easy 2D or 3D visualization, making it straightforward to inspect cluster patterns.
#In high-dimensional space, visualizing clusters is impossible; PCA provides a clear view of cluster distribution.

#Slight improvement in clustering metrics
#Metrics like Silhouette Score usually increase, while Davies-Bouldin Index decreases after PCA, indicating improved clustering quality.
#This confirms that dimensionality reduction can help the clustering algorithm focus on meaningful structure rather than noise.

#Reduced computational complexity
#Fewer dimensions mean faster computation for K-Means and other clustering algorithms, especially on large datasets.
#Memory usage and runtime improve without significant loss of information.

#Minimal impact on cluster assignments
#While PCA changes the feature space, the overall grouping pattern of clusters often remains similar.
#PCA mainly refines the cluster structure rather than creating completely different clusters.
```

```python
#PCA before clustering is beneficial for high-dimensional datasets: it reduces noise, improves cluster separation, speeds up computation, and enables visualization
```

```python
Recommendations for Using PCA and Clustering
1. When to Use Clustering on Original Data

Use when interpretability is important: you want to understand how the original features contribute to cluster formation.

Suitable for low to moderately high-dimensional datasets where computational cost is manageable.

Helpful when all features are meaningful and you do not want to lose any information.

2. When to Use PCA Before Clustering

Use when the dataset is high-dimensional, noisy, or has correlated features.
#PCA reduces dimensions, making clustering faster and more efficient.
#Enhances cluster compactness and separation, improving clustering metrics like Silhouette Score and Davies-Bouldin Index.
#Essential for visualizing clusters in 2D or 3D.

#3. When to Use Both Techniques Together
#Use PCA as a preprocessing step before clustering on high-dimensional data:
#Reduces noise and redundancy.
#Improves clustering performance and computational efficiency.
#Enables clear visualization while still capturing the main variance of the data.
#Recommended for exploratory data analysis, pattern recognition, and situations where speed and clarity are needed without losing too much information.

#4. Trade-Off Awareness
#PCA improves efficiency and visualization but may obscure direct feature interpretation.
#Clustering on original features preserves interpretability but may be slower and affected by irrelevant/noisy features.
```
