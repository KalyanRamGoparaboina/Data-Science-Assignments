```python
#Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
import numpy as np
def is_prime(number):
    if number <= 1:
        return False 
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False  
    return True 
    #The program determines whether a given number is prime by checking if it has any divisors other than 1 and itself
```

```python
#Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.
import random

def generate_and_check_product():
    """Generates two random numbers, asks the user for the product, and checks the answer."""
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    correct_product = num1 * num2

    user_answer_str = input(f"What is the product of {num1} and {num2}? ")

    try:
        user_answer = int(user_answer_str)
        if user_answer == correct_product:
            print("Correct! Well done!")
        else:
            print(f"Incorrect. The correct answer is {correct_product}.")
    except ValueError:
        print("Invalid input. Please enter an integer.")
#The program tests the user's multiplication skills by generating random numbers and verifying the correctness of their response
```

```python
#Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

number_type = 'odd'

print(f"Printing squares of {number_type} numbers between 100 and 200:")

for number in range(100, 201):
    if number_type == 'even':
        if number % 2 == 0:
            square = number ** 2
            print(f"{number}^2 = {square}")
    elif number_type == 'odd':
        if number % 2 != 0:
            square = number ** 2
            print(f"{number}^2 = {square}")
    else:
        print("Invalid number_type specified.")
        break
#The program successfully prints the squares of all chosen (even/odd) numbers between 100 and 200, demonstrating the use of loops and conditional logic in Python
```

```python
# Program to count the number of words in a given text

input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Split the text into words
words = input_text.split()

# Create a dictionary to store word counts
word_count = {}

# Count each word
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Display results
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

```python
import string

def is_palindrome(text: str) -> bool:
    # Remove spaces and punctuation, and convert to lowercase
    cleaned_text = ''.join(ch.lower() for ch in text if ch.isalnum())
    
    # Check if cleaned_text is equal to its reverse
    return cleaned_text == cleaned_text[::-1]
```
