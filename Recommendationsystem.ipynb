```python
df = df.dropna(subset=['name', 'genre']).copy()  # <-- add .copy()

if 'type' in df.columns:
    most_common_type = df['type'].mode()[0]
    df.loc[:, 'type'] = df['type'].fillna(most_common_type)

if 'rating' in df.columns:
    df.loc[:, 'rating'] = df['rating'].fillna(df['rating'].mean())

df.loc[:, 'episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
df.loc[:, 'episodes'] = df['episodes'].fillna(df['episodes'].median())
```

```python
df = df.drop_duplicates()
df.isnull().sum()
```

```python
#feature extraction
```

```python
feature=df.copy()
```

```python
df_genre = feature['genre'].str.get_dummies(sep=',')
df_genre.head()
```

```python
df_feature = pd.concat([df_genre, feature[['rating', 'members']]], axis=1)
df_feature.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_feature[['rating', 'members']] = scaler.fit_transform(df_feature[['rating', 'members']])
df_feature.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(df_feature)
cosine_sim.shape
```

```python
def recommend_anime(title, df, df_feature, top_n=5, threshold=0.5):
    df_reset = df.reset_index(drop=True)
    df_reset['name_lower'] = df_reset['name'].str.lower()
    title_lower = title.lower()
    if title_lower not in df_reset['name_lower'].values:
        print(f"âŒ '{title}' not found in dataset!")
        return []
    idx = df_reset[df_reset['name_lower'] == title_lower].index[0]
    sim_scores = cosine_similarity(df_feature, df_feature.iloc[[idx]]).flatten()
    sim_scores_filtered = [(i, score) for i, score in enumerate(sim_scores)
                           if score >= threshold and i != idx]
    top_similar = sorted(sim_scores_filtered, key=lambda x: x[1], reverse=True)[:top_n]
    recommendations = [(df_reset.iloc[i]['name'], round(score, 3)) for i, score in top_similar]
    print(f"\nðŸŽ¬ Recommendations for '{title}' (threshold={threshold}):")
    for name, score in recommendations:
        print(f"  - {name} (similarity: {score})")
    return recommendations
recommend_anime("Naruto", df, df_feature, top_n=5, threshold=0.3)
```

```python
#Evaluation
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, df_train, df_test = train_test_split(
    df_feature, df,
    test_size=0.2,
    random_state=42
)
X_train.shape, X_test.shape
```

```python
def evaluate_recommendations(df, df_feature, top_n=5, thresholds=[0.3, 0.4]):
    df_reset = df.reset_index(drop=True)
    results = {}

    for threshold in thresholds:
        precisions = []
        recalls = []

        for title in df_reset['name'][:50]:  # limit to 50 for speed
            recs = recommend_anime(title, df_reset, df_feature, top_n=top_n, threshold=threshold)
            if not recs:
                continue

            recommended_names = [r[0] for r in recs]
            # Assume relevance = similarity > 0.7 (for demo)
            relevant_items = [r for r in recs if r[1] > 0.7]

            precision = len(relevant_items) / len(recs) if recs else 0
            recall = len(relevant_items) / len(recommended_names) if recommended_names else 0

            precisions.append(precision)
            recalls.append(recall)

        results[threshold] = {
            "avg_precision": round(sum(precisions) / len(precisions), 3) if precisions else 0,
            "avg_recall": round(sum(recalls) / len(recalls), 3) if recalls else 0
        }

    return results
    thresholds_to_test = [0.1, 0.2, 0.3, 0.4]
metrics = evaluate_recommendations(df, df_feature, top_n=5, thresholds=thresholds_to_test)
```

```python
print("\nðŸ“Š Evaluation Metrics:")
for th, vals in metrics.items():
    print(f"Threshold: {th}")
    print(f"  Avg Precision: {vals['avg_precision']}")
    print(f"  Avg Recall: {vals['avg_recall']}")
```
