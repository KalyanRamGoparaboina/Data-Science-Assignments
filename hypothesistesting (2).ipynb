```python
import numpy as np
from scipy.stats import t
```

```python
#given data
sample_mean=3050
theoritical_meean=1000+5*600
standard_deviation=5*25
sample_size=25
alpha=0.05
```

```python
#TASK 1:State The Hpothesis
#A hypothesis is a statement or assumption that we make about a population or relationship, which we then test using data.In statistics, hypotheses are used to check whether what we observe in the sample data provides enough evidence to make conclusions about the whole population
#It consists of two types:
#1.Null Hypthesis
#2.Alternate Hypothesis
```

```python
# TASK2:calculating t_statistics
standard_error=standard_deviation/np.sqrt(sample_size)
t_statistics=(sample_mean-theoritical_meean)/standard_error
```

```python
print("Standard Error",standard_error)
print("T statistics",t_statistics)
```

```python
# TASK3:Determine the critical value
dof=sample_size-1
critical_value=t.ppf(1-alpha,dof)
```

```python
critical_value
```

```python
#TASK4:MAKE A DECISION
if t_statistics<critical_value:
    print("Accept H0:The operating cost for a franchise in a week is not changed and remained same as $1000+$5X for X=600")
else:
    Print("Reject H0:The operating cost for a franchise in a week is changed as greater than $1000+$5X for X=600")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
#Task1:State the hypothesis
#H0:There is no significant association between the type of smart home device purchased(smart thermostats vs. smart light) and the customer satisfaction level
#H1:There is significant association between the type of smart home device purchased(smart thermostats vs. smart light) and the customer satisfaction level.
```

```python
#TASK 2:Compute the chi square statstic
data=np.array([[50,70],
             [80,100],
             [60,90],
             [30,50],
             [20,50]])
df=pd.DataFrame(data,columns=["Smart Thermostat","Smart Light"],
               index=["Very Statisfied","Satisied","Neutarl","Unsatisfied","Very satisfied"])
df
```

```python
chi2_stats,p_val,dof,expected=chi2_contingency(df)
```

```python
print("chi-square Statistics:",chi2_stats)
print("Degree Of Freedom:",dof)
print("P-Value",p_val)
print("Expected Frequency:\n",expected)
```

```python
#Task3:DETERMINE THE CRITICAL VALUE
from scipy.stats import chi2
alpha=0.05
critical_value=chi2.ppf(1-alpha,dof)
print("Critical Value:",critical_value)
```

```python
#Task 4:Make the decision
if chi2_stats<critical_value:
    print("Accept H0")
    print("There is no significant association between the type of smart home device purchased(smart thermostats vs. smart light) and the customer satisfaction level")
else:
    print("Reject H0")
    print("There is significant association between the type of smart home device purchased(smart thermostats vs. smart light) and the customer satisfaction level")
```
